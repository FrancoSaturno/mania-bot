```python
import numpy as np
from mss import mss
import time
import cv2
```

```python
def frame_for_circles():
    with mss() as sct:
        playfield = {"top": 0, 
                     "left": 240, 
                     "width": 1440, 
                     "height": 1080}
        raw_frame = sct.grab(playfield)
        osu_frame_np = np.array(raw_frame)
        osu_frame_gray_np = cv2.cvtColor(osu_frame_np, cv2.COLOR_BGRA2GRAY)
        return osu_frame_gray_np
    
```
