```python
import cv2
import numpy as np
import time
from mss import mss
from matplotlib import pyplot as plt

```

```python
# ============================================================================
# CIRCLE DETECTOR CLASS (copy this into your environment later)
# ============================================================================
class CircleDetector:
    def __init__(self):
        """Initialize the blob detector with osu!-specific parameters"""
        params = cv2.SimpleBlobDetector_Params()
        
        # Enable area filtering - VERY PERMISSIVE
        params.filterByArea = True
        params.minArea = 35       # LOWER - catch tiny blobs
        params.maxArea = 1000     # HIGHER - catch huge blobs
        
        # DISABLE all shape filtering
        params.filterByCircularity = False
        params.filterByConvexity = False
        params.filterByInertia = False
        
        # Enable color filtering - only detect BRIGHT blobs
        params.filterByColor = True
        params.blobColor = 255    # Only detect white/bright blobs
        
        # Threshold settings - LOWER to detect dimmer circles
        params.minThreshold = 5   # Start detecting at brightness 5/255
        params.maxThreshold = 255  # Go up to full white
        params.thresholdStep = 5  # Check every 5 brightness levels
        
        self.blob_detector = cv2.SimpleBlobDetector_create(params)
    
    def detect_bright_circles(self, frame_84x84):
        """
        Detects bright circles using SimpleBlobDetector
        Returns top 2 brightest circles sorted by brightness
        
        Args:
            frame_84x84: numpy array of shape (84, 84) with uint8 values
        
        Returns:
            List of dicts: [{"x": 0.5, "y": 0.3, "brightness": 0.9}, ...]
        """
        # Run blob detection on the input frame
        keypoints = self.blob_detector.detect(frame_84x84)
        
        # Convert keypoints to our format with brightness
        circles = []
        for kp in keypoints:
            # Extract center position
            x, y = kp.pt
            x_int, y_int = int(x), int(y)
            
            # Calculate radius for brightness sampling
            radius = int(kp.size / 2)
            
            # Calculate bounding box (with boundary checks)
            y1, y2 = max(0, y_int-radius), min(84, y_int+radius)
            x1, x2 = max(0, x_int-radius), min(84, x_int+radius)
            
            # Calculate average brightness in the blob region
            brightness = frame_84x84[y1:y2, x1:x2].mean() / 255.0
            
            # Store normalized coordinates and brightness
            circles.append({
                "x": x / 84.0,           # Normalize to 0-1 range
                "y": y / 84.0,           # Normalize to 0-1 range
                "brightness": brightness  # Already 0-1 range
            })
        
        # Sort by brightness (brightest first), return top 2
        circles.sort(key=lambda c: c["brightness"], reverse=True)
        return circles[:2]  # Changed from [:3] to [:2]


# ============================================================================
# TEST SETUP
# ============================================================================
def capture_osu_frame():
    """
    Captures a frame from your osu! screen region
    Adjust these coordinates to match your setup!
    """
    # ADJUST THESE TO YOUR OSU! WINDOW
    # Example: if osu! is at x=240, y=0, width=1440, height=1080
    monitor = {
        "top": 0,      # Change this to your osu! window Y position
        "left": 240,   # Change this to your osu! window X position
        "width": 1440, # Change this to your osu! window width
        "height": 1080 # Change this to your osu! window height
    }
    
    with mss() as sct:
        # Capture the screen
        screenshot = sct.grab(monitor)
        frame = np.array(screenshot)
        
        # Convert BGRA to grayscale
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGRA2GRAY)
        
        # Resize to 84x84 (what your agent sees)
        frame_84x84 = cv2.resize(frame_gray, (84, 84), interpolation=cv2.INTER_AREA)
        
        return frame_84x84, frame_gray  # Return both for visualization


def visualize_with_matplotlib(frame_84x84, circles, frame_count):
    """
    Shows the detection results using matplotlib (displays in terminal/notebook)
    """
    # Create figure
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    
    # Show grayscale frame
    ax.imshow(frame_84x84, cmap='gray', vmin=0, vmax=255)
    
    # Draw detected circles
    for i, circle in enumerate(circles):
        x_px = circle['x'] * 84
        y_px = circle['y'] * 84
        
        # Color: Red for brightest, yellow for 2nd
        colors = ['red', 'yellow']  # Changed from ['red', 'yellow', 'cyan']
        color = colors[i] if i < len(colors) else 'white'
        
        # Draw circle
        circle_patch = plt.Circle((x_px, y_px), 8, color=color, fill=False, linewidth=2)
        ax.add_patch(circle_patch)
        
        # Draw brightness text
        ax.text(x_px + 10, y_px, f"{circle['brightness']:.2f}", 
                color=color, fontsize=10, weight='bold')
    
    ax.set_title(f'Frame {frame_count} - Detected {len(circles)} circles')
    ax.axis('off')
    
    plt.tight_layout()
    plt.show()


# ============================================================================
# MAIN TEST LOOP
# ============================================================================
def main():
    print("Starting circle detection test...")
    print("Make sure osu! is running and visible!")
    
    # ========== 10 SECOND COUNTDOWN ==========
    print("\nStarting in:")
    for i in range(10, 0, -1):
        print(f"  {i}...")
        time.sleep(1)
    print("GO!\n")
    # =========================================
    
    # Initialize detector
    detector = CircleDetector()
    
    # Performance tracking
    frame_times = []
    detect_times = []
    
    try:
        frame_count = 0
        while True:
            frame_start = time.perf_counter()
            
            # Capture frame
            frame_84x84, frame_full = capture_osu_frame()
            
            # Detect circles
            detect_start = time.perf_counter()
            circles = detector.detect_bright_circles(frame_84x84)
            detect_time = (time.perf_counter() - detect_start) * 1000  # Convert to ms
            
            # Track performance
            frame_time = (time.perf_counter() - frame_start) * 1000
            frame_times.append(frame_time)
            detect_times.append(detect_time)
            
            # Print results every frame
            avg_frame_time = np.mean(frame_times[-10:]) if len(frame_times) >= 10 else np.mean(frame_times)
            avg_detect_time = np.mean(detect_times[-10:]) if len(detect_times) >= 10 else np.mean(detect_times)
            
            print(f"\n--- Frame {frame_count} ---")
            print(f"Total time: {avg_frame_time:.2f}ms (capture + detect)")
            print(f"Detection time: {avg_detect_time:.2f}ms")
            print(f"Detected {len(circles)} circles:")
            
            for i, circle in enumerate(circles):
                print(f"  Circle {i+1}: x={circle['x']:.3f}, y={circle['y']:.3f}, brightness={circle['brightness']:.3f}")
            
            # Show visualization every 10 frames using matplotlib
            if frame_count % 10 == 0:
                visualize_with_matplotlib(frame_84x84, circles, frame_count)
            
            frame_count += 1
            
            # Stop after 100 frames for testing (remove this if you want infinite loop)
            if frame_count >= 100:
                print("\nReached 100 frames, stopping...")
                break
    
    except KeyboardInterrupt:
        print("\n\nStopping test...")
    
    finally:
        # Print final statistics
        if frame_times:
            print(f"\n=== Performance Summary ===")
            print(f"Frames processed: {len(frame_times)}")
            print(f"Avg total time: {np.mean(frame_times):.2f}ms")
            print(f"Avg detection time: {np.mean(detect_times):.2f}ms")
            print(f"Detection overhead: {np.mean(detect_times) / np.mean(frame_times) * 100:.1f}%")


if __name__ == "__main__":
    main()
```
